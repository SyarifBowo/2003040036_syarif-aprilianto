```python
# mengimpor pustaka Geopandas ke dalam program Python Anda
import geopandas as gpd
```

```python
# membaca file GeoJSON yang disebut "RBI25K_ADMINISTRASI_LN.json".
gdf = gpd.read_file("RBI25K_ADMINISTRASI_LN.json")

# menampilkan lima baris pertama dari GeoDataFrame
gdf.head()
```

```python
# memfilter kolom yang akan di panggil
# Setelah pemilihan kolom, fungsi fillna(0) digunakan untuk menggantikan nilai-nilai yang kosong (NaN)
# dalam GeoDataFrame dengan nilai 0.
gdf = gdf[['NAMOBJ','FCODE','REMARK','METADATA', 'geometry']].fillna(0)
```

```python
# menginstal tiga pustaka Python, yaitu Folium, Matplotlib, dan Mapclassify,
# menggunakan perintah pip langsung dari dalam sebuah lingkungan notebook atau console.
!pip install folium matplotlib mapclassify
```

```python
# mengimpor pustaka Folium ke dalam program Python Anda.
import folium

# Membuat objek peta interaktif
m = gdf.explore(column="METADATA", name="MetaData")

# Menambahkan kontrol lapisan ke peta
folium.LayerControl().add_to(m)

# Menampilkan peta interaktif
m
```
